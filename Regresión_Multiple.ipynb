```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('cuentas.xlsx')
```

```python
df=df[['precio','enganche','descuento','semana','monto_financiado','costo_total','monto_accesorios','status','fraude','inversion','pagos_realizados','reautorizacion','porc_enganche','score_buro']]
```

```python
df.isnull().sum()
```

```python
df['porc_enganche']=df['porc_enganche'].fillna(0)
```

```python
Corr_Factors=df.corr()
Corr_Factors

Corr_Factors1=abs(Corr_Factors)
Corr_Factors1

Heat_Map= sns.heatmap(Corr_Factors1, cmap = 'YlGnBu', annot=True, fmt=".2f")
Heat_Map
```

Dependiente Enganche

```python
Vars_Indep= df[['precio','monto_financiado']]
Var_Dep= df['enganche']
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
type(model)
```

```python
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
model.__dict__
```

Modelo:
y=260257.00754878x1 + -0.98847381x2 + 3.4037533329509415

```python
model.score(Vars_Indep,Var_Dep)
```

```python
y_pred= model.predict(X=df[['precio','monto_financiado']])
y_pred
```

```python
sns.scatterplot(x='precio', y='enganche', color="blue", data=df)
```

```python
sns.scatterplot(x='monto_financiado', y='precio', color="blue", data=df)
```

```python
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
coef_Deter
```

```python
coef_Correl=np.sqrt(coef_Deter)
coef_Correl
```
